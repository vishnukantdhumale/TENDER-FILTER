```python
import pandas as pd
```

```python
#reading tender table from site using pandas
url="https://sourcing.amns.in/E_Tender/USER/Tender_Download.aspx?id=1"
tables=pd.read_html(url)
#10 table are there out of which table number 8 is rquired tender table
df=tables[8]
```

```python
#sample dataframe
df.head(2)
```

```python
#Rename all the columns in dataframe
df.rename(columns={df.columns[0]: 'Sr.No.'}, inplace=True)
df.rename(columns={df.columns[1]: 'Tender_Date'}, inplace=True)
df.rename(columns={df.columns[2]: 'Tender_No'}, inplace=True)
df.rename(columns={df.columns[3]: 'Tender Description'}, inplace=True)
df.rename(columns={df.columns[4]: 'Last_Date'}, inplace=True)
df.rename(columns={df.columns[5]: 'RFQ_NO'}, inplace=True)
df.head(2)




```

```python
#dropping row 1 having index 0 because it is same as column table
df.drop(index=[0], inplace=True)
```

```python
#sample data after rename and drop of row
df.head(2)
```

```python
#Filter rows having word "FILTER"
df.dropna(subset=['Tender Description'], inplace=True)
column_to_search = 'Tender Description'

# Define the word you want to search for
word_to_search = 'FILTER'

# Filter the DataFrame based on whether the specified column contains the word
filtered_df = df[df[column_to_search].str.contains(word_to_search, case=False)]

# Display the filtered DataFrame
print(filtered_df)
```

```python
#Filter word having "FAN"
word_to_FIND = 'FAN'

# Filter the DataFrame based on whether the specified column contains the word
filtered_df1 = df[df[column_to_search].str.contains(word_to_FIND, case=False)]

# Display the filtered DataFrame
print(filtered_df1)
```

```python
#Join two different dataset to from one dataset formed by filtering word fan and filter
result = pd.concat([filtered_df, filtered_df1], axis=0)
result
```
